```python
import re
def clean_ocr_text(raw_text):
    # Remove extra whitespace and newlines
    cleaned = re.sub(r'\n+', ' ', raw_text)
    # Remove special characters that are OCR artifacts
    cleaned = re.sub(r'[^\w\s\-.,]', '', cleaned)
    # Remove extra spaces
    cleaned = ' '.join(cleaned.split())
    return cleaned.strip()
```

```python
from paddleocr import PaddleOCR
import os

# Initialize
ocr = PaddleOCR(use_textline_orientation=True, lang='en')

# Replace 'test.jpg' with your actual image file name
image_path = 'How to Add FingerPrint/1.jpg'

# Check if file exists
if not os.path.exists(image_path):
    print(f"Error: Image file '{image_path}' not found!")
    exit()

try:
    result = ocr.predict(image_path)
    
    # Debug: Print the structure of result
    print(f"Result type: {type(result)}")
    print(f"Result length: {len(result) if result else 0}")
    
    # Fixed text extraction
    print("\nRESULTS:")
    if result and len(result) > 0 and result[0]:
        print(f"Found {len(result[0])} text regions")
        
        for i, line in enumerate(result[0]):
            try:
                # More robust extraction
                if len(line) >= 2 and len(line[1]) >= 1:
                    text = line[1][0]  # Get the full text
                    confidence = line[1][1] if len(line[1]) >= 2 else "N/A"
                    print(f"{i+1}. Text: '{text}' (Confidence: {confidence})")
                else:
                    print(f"{i+1}. Invalid line structure: {line}")
            except Exception as e:
                print(f"{i+1}. Error processing line: {e}")
    else:
        print("No text found or empty result")
    
    # Alternative: Just print all text in one go
    print("\nAll recognized text:")
    if result and len(result) > 0 and result[0]:
        try:
            all_text = []
            for line in result[0]:
                if len(line) >= 2 and len(line[1]) >= 1:
                    all_text.append(line[1][0])
            
            if all_text:
                print(" ".join(all_text))
            else:
                print("No valid text extracted")
        except Exception as e:
            print(f"Error extracting all text: {e}")
    else:
        print("No text to display")

except Exception as e:
    print(f"An error occurred: {e}")
    print("This might be due to:")
    print("1. Image file not found")
    print("2. Corrupted image file")
    print("3. PaddleOCR installation issues")
    print("4. Insufficient memory")
```

```python
from paddleocr import PaddleOCR

# Initialize with better settings for text detection
ocr = PaddleOCR(
    use_angle_cls=True,  # Enable angle classification
    lang='en',
    det_db_thresh=0.3,   # Lower detection threshold
    det_db_box_thresh=0.5,  # Box threshold
    rec_batch_num=6      # Batch processing
)

# Replace with your actual image file name
image_path = 'How to Add FingerPrint/1.jpg'
result = ocr(image_path, cls=True)  # Direct call to ocr object

# Method 1: Print each detected text element
print("Individual text elements:")
if result[0]:
    for line in result[0]:
        text = line[1][0].strip()  # Extract text and remove whitespace
        if len(text) > 1:  # Only print if more than 1 character
            print(text)
else:
    print("No text found")

print("\n" + "="*50 + "\n")

# Method 2: Combine all text into one string
print("Combined text:")
if result[0]:
    all_text = ' '.join([line[1][0].strip() for line in result[0] if len(line[1][0].strip()) > 1])
    print(all_text)
else:
    print("No text found")

print("\n" + "="*50 + "\n")

# Method 3: Sort by position and combine (reads left-to-right, top-to-bottom)
print("Text sorted by position:")
if result[0]:
    # Sort by y-coordinate (top to bottom), then x-coordinate (left to right)
    sorted_results = sorted(result[0], key=lambda x: (x[0][0][1], x[0][0][0]))
    sorted_text = ' '.join([line[1][0].strip() for line in sorted_results if len(line[1][0].strip()) > 1])
    print(sorted_text)
else:
    print("No text found")
```

```python
def icon_detection(img_path, temp_img_path):
    import cv2
    import numpy as np
    
    # Step 1 : Read the image
    img = cv2.imread(img_path)
    temp_img = cv2.imread(temp_img_path)

    # Step 2 : Convert the image into grayscale image
    img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    temp_img_gray = cv2.cvtColor(temp_img, cv2.COLOR_BGR2GRAY)

    # Step 3 : Get template dimensions
    temp_height, temp_width = temp_img_gray.shape

    # Step 4 : Check if the template is smaller than the main image
    if temp_img_gray.shape[0] > img_gray.shape[0] or temp_img_gray.shape[1] > img_gray.shape[1]:
        print("Error : Template is larger than the main image")
        exit()

    # Step 5 : Perform template matching
    result = cv2.matchTemplate(img_gray, temp_img_gray, cv2.TM_CCOEFF_NORMED)
    threshold = 0.95
    loc = np.where(result >= threshold)

    # Step 6 : Store bounding box information
    bounding_boxes = []

    for point in zip(*loc[::-1]):
        x, y = point
        width = temp_width
        height = temp_height

        bbox = {
        'x': x,
        'y': y, 
        'width': width,
        'height': height,
        'x2': x + width,  # bottom-right x coordinate
        'y2': y + height  # bottom-right y coordinate
        }
        bounding_boxes.append(bbox)

    # Step 7 : Draw rectangle on image
    cv2.rectangle(img, (x, y), (x + width, y + height), (0, 255, 255), 2)

    print(f"\nTotal matches found: {len(bounding_boxes)}")

    if bounding_boxes:
        print(f"\nFirst match details:")
        first_bbox = bounding_boxes[0]
        print(f"X: {first_bbox['x']}")
        print(f"Y: {first_bbox['y']}")
        print(f"Width: {first_bbox['width']}")
        print(f"Height: {first_bbox['height']}")

    cv2.imshow("Detected", img)

    cv2.waitKey(0)
    cv2.destroyAllWindows()
```

```python
def instruction_match(instruction_dict, ocr_text, threshold_value):
    """
    Matches OCR text to instruction keywords using fuzzy matching.

    Args:
    - instruction_dict: A dictionary of instructions -> keyword
    - ocr_text: A string containing text extracted by OCR
    - threshold_value: Minimum match score (0-100) to be considered a match

    Returns:
    - A list of (instruction, matched_keyword, score) sorted by score in descending order
    """
    from rapidfuzz import fuzz
    matched_instructions = list()
    #ocr_text_lower = ocr_text.lower()
    for instruction, keyword in instruction_dict.items():
        score = fuzz.partial_ratio(keyword.lower(), ocr_text)
        if score >= threshold_value:
            matched_instructions.append((keyword, ocr_text, score))

    # Sort by score (highest first)
    matched_instructions.sort(key=lambda x: x[2], reverse=True)
    return matched_instructions
```

```python
# this function draws bounding boxes for the ocr extracted text in the image

def draw_bbox_ocr(img_path, matched_texts):
    import easyocr
    import cv2

    reader = easyocr.Reader(["en"], gpu=False)
    img = cv2.imread(img_path)

    # Perform OCR
    results = reader.readtext(img)

    boxes = []
    for (b_box, text, confidence) in results:
        for matched_text, _ in matched_texts:
            if text.strip().lower() == matched_text.strip().lower():
                (top_left, top_right, bottom_right, bottom_left) = b_box

                x = int(top_left[0])
                y = int(top_left[1])
                w = int(top_right[0] - top_left[0])
                h = int(bottom_left[1] - top_left[1])

                cv2.rectangle(img, (x, y), (x + w, y + h), (0, 255, 255), 2)
                cv2.putText(img, text, (x, y - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (36, 255, 12), 1)

                boxes.append({
                    "text" : text,
                    "x" : x,
                    "y" : y,
                    "width" : w,
                    "height" : h
                })


    cv2.imwrite("output_with_boxes.jpg", img)

    cv2.waitKey(0)
    cv2.destroyAllWindows()
    return boxes
```
